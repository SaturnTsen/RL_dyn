```python
import torch
```

```python
import matplotlib.pyplot as plt
rewards = []

# glob
r = range(25, 186)
for ep in r:
    data = torch.load(f'checkpoint_ep{ep}.pth', weights_only=False)
    rewards.append(data['reward'])
print(rewards)
```

```python
import matplotlib.pyplot as plt
plt.plot(r, rewards)
# 5 moving average
window_l = 5
window_r = 5
moving_avg = []
for i in range(len(rewards)):
    start = max(0, i - window_l)
    end = min(len(rewards), i + window_r + 1)
    moving_avg.append(sum(rewards[start:end]) / (end - start))
plt.plot(r, moving_avg, label='Moving Average (window=5)')
plt.title('Reward over Episodes')
plt.xlabel('Episode') 
```
