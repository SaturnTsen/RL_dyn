```python
import torch
```

```python
import matplotlib.pyplot as plt
rewards = []

# glob
r = range(25, 170)
for ep in r:
    data = torch.load(f'checkpoint_ep{ep}.pth', weights_only=False)
    rewards.append(data['reward'])
print(rewards)
```

```python
import matplotlib.pyplot as plt
plt.plot(r, rewards)
# 5 moving average
window_size = 15
moving_avg = [sum(rewards[i:i+window_size])/window_size for i in range(len(rewards)-window_size+1)]
plt.plot(r[:len(moving_avg)], moving_avg, label='Moving Average (window=15)')
plt.title('Reward over Episodes')
plt.xlabel('Episode') 
```
